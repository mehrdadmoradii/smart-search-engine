```python
from elasticsearch import Elasticsearch
```

```python
client = Elasticsearch('http://localhost:9200')
```

<h1>Elasticsearch match query</h1>

```python
query = 'How many people have died during Black Death?'
```

```python
res = client.search(
    index='articles',
    query={
        "match": {
            "text": query
        }
    }
)['hits']['hits']

for r in res:
    r = r['_source']
    print(r['source_url'])
    print(r['text'][:200], end='\n\n\n')
```

<h1>Elasticsearch term query</h1>

```python
res = client.search(
    index='articles',
    query={
        'term': {
            'article_title': 'Swine influenza'
        }
    }
)['hits']['hits']

for r in res:
    r = r['_source']
    print(r['source_url'])
    print(r['text'][:200], end='\n\n\n')
```

<h1>Elasticsearch boolean query</h1>

```python
res = client.search(
    index='articles',
    body={
      "size": 20,
      "query": {
        "bool": {
          "must": [
            {"multi_match": {
              "query": "*World Health Organization",
              "fields": ["text^3", "article_title^3"]
            }}
          ],
          "must_not": [
            {"match": {"main_section": "External links"}}
          ],
          "should": [
            {"match": {"main_section": {"query": "Summary","boost": 2}}}
          ]
        }
      },
      "highlight": {
        "fields": {
          "text": {}
         } 
      }
    }
)['hits']['hits']


for r in res:
    print(r['_source']['text'][:200])
    print(r['highlight']['text'])
    print('\n\n')
```
